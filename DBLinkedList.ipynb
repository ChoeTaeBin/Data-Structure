dummyNode를 두어 코드를 간단하게 수정한 양방향 LinkedList

```python
class DBLinkedList:


  class Node:#노드
    def __init__(self,data = None, prev = None, next = None):
      self.data = data
      self.prev = prev #전의 것
      self.next = next #뒤의 것

  def __init__(self):
    #dummyNode 생성
    self.head = self.Node()
    self.tail = self.Node(None,self.head,None)
    self.head.next = self.tail

    self.len = 0 #요소의 갯수를 나타냄


  def isEmpty(self):#비어있는지
    return self.len == 0


  def append(self,data):#뒤에 data를 추가
    nNode = self.Node(data,self.tail.prev,self.tail)
    
    self.tail.prev.next = nNode #앞의 것과 연결
    self.tail.prev = nNode#뒤의 것과 연결
  
    self.len += 1


  def insert(self,idx,data):#원하는 위치에 삽입

    if(idx<0):#음수 idx변환
      idx += self.len

    if(idx< 0):#그래도 idx가 0보다 작으면  맨 앞으로 삽입
      idx = 0
    elif(self.len<idx):#마지막 인덱스 보다 크면 마지막으로 간주
      idx = self.len

    #효율을 위해 앞,뒤에서 접근하는 case를 나눔
    if(idx <= self.len/2): #앞에서 부터 가는 것이 빠를 때
      #앞에서 부터 idx-1번 까지 이동
      current = self.head
      for i in range(idx):
        current = current.next#이동
      print("앞에서 부터: ", idx, "이동")
    else: #뒤에서 접근 하는게 빠를때
      current = self.tail
      for i in range(self.len-idx+1):
        current = current.prev
      print("뒤에서 부터: ", self.len-idx+1, "이동")

    #삽입
    nNode = self.Node(data,current,current.next)#새 노드 생성
    nNode.prev.next = nNode #앞 연결
    nNode.next.pre = nNode #뒤 연결
    
    self.len+=1 #길이 증가

  def delete(self,idx):#idx위치의 노드를 삭제
    if self.isEmpty():#아무 것도 할 것이 없음
      return

    if(idx<0):#음수 idx변환
      idx += self.len

    if(idx< 0):#그래도 idx가 0보다 작으면 0으로 간주
      idx = 0
    elif(self.len-1<idx):#마지막 인덱스 보다 크면 마지막으로 간주
      idx = self.len-1

    #idx-1번 까지 이동
    current = self.head
    for i in range(idx):
     current = current.next#이동
    
    #삭제
    current.next = current.next.next
    current.next.prev = current

    self.len-=1

  def printAll(self):#모두 출력
    if self.isEmpty():#리스트가 비어있다면
      print("list is empty")
      return

    print("[start",end=" -> ")
    current = self.head.next
    while current != self.tail:
      print(current.data, end = " -> ")
      current = current.next
    print("end]")


  def search(self,data):#값이 data인 idx 반환 없으면 -1 반환
    if self.isEmpty():
      return -1
    
    idx = 0
    current = self.head.next
    while current != self.tail:
      if(current.data == data):
        return idx
      current = current.next
      idx+=1
    return -1


  def getData(self,idx): #[]연산자 오버로딩의 효과

    if self.isEmpty():#리스트가 비어있다면
      print("list is empty")
      return

    if(idx<0):#음수 idx변환
      idx += self.len

    if(idx< 0):#그래도 idx가 0보다 작으면 0으로 간주
      idx = 0
    elif(self.len-1<idx):#마지막 인덱스 보다 크면 마지막으로 간주
      idx = self.len-1
    
    current = self.head.next
    for i in range(idx):
      current = current.next
    return current.data


  def swap(self,idx1,idx2):#idx1의 데이터와 idx2의 데이터를 바꿈

    if self.isEmpty():#리스트가 비어있다면
      print("list is empty")
      return
    
    Node1 = None
    Node2 = None

    #이상한 idx들 처리
    if(idx1<0):#음수 idx1변환
      idx1 += self.len

    if(idx1< 0):#그래도 idx1가 0보다 작으면 0으로 간주
      idx1 = 0
    elif(self.len-1<idx1):#마지막 인덱스 보다 크면 마지막으로 간주
      idx1 = self.len-1

    if(idx2<0):#음수 idx1변환
      idx2 += self.len

    if(idx2< 0):#그래도 idx2가 0보다 작으면 0으로 간주
      idx2 = 0
    elif(self.len-1<idx2):#마지막 인덱스 보다 크면 마지막으로 간주
      idx2 = self.len-1
      
    #idx1노드 찾기
    Node1 = self.head.next
    for i in range(idx1):
      Node1 = Node1.next
    
    #idx2노드 찾기
    Node2 = self.head.next
    for i in range(idx2):
      Node2 = Node2.next

    #swap
    Node1.data,Node2.data = Node2.data,Node1.data


  def reverse(self):#배열을 뒤집는다.
    for i in range(int(self.len/2)):
      self.swap(i,self.len-1-i)
```
