```python
class SortLab:
  
  def bubbleSort(self,arr):
    numOper = 0
    for end in range(len(arr)-1,0,-1):
      for i in range(0,end,1):
        numOper += 1
        if(arr[i] > arr[i+1]):
          arr[i],arr[i+1] = arr[i+1],arr[i]
    return numOper
  
  def selectionSort(self,arr):
    for start in range(len(arr)-1):
      localMinIdx = start#구간에서 제일 작은값의 인덱스
      for i in range(start+1,len(arr)):
        if(arr[i] < arr[localMinIdx]):
          localMinIdx = i
      #print("localMin: ",localMinIdx, "번의 ", arr[localMinIdx])
      arr[start],arr[localMinIdx] = arr[localMinIdx],arr[start]#가장 작은 값을 왼쪽에 위치시킴

  def insertionSort(self,arr):
    for cur in range(1,len(arr)):
      curData = arr[cur] #복사 해둠

      i = cur-1
      while(i>-1):
        if(curData<arr[i]):
          arr[i+1] = arr[i] #한칸씩 밀기
        else:
          break
        i-=1
      arr[i+1] = curData
     
  def quickSort(self, arr, left, right):

    if(left>=right):#종결 조건
      return 0
    else:
      pivotIdx, numOper = self.partition(arr,left,right)

      numOper += self.quickSort(arr,left,pivotIdx-1)#왼쪽 정렬
      numOper += self.quickSort(arr,pivotIdx+1,right)#오른쪽 정렬

      return numOper

  def selectPivot(self,arr,left,right):#계산수를 줄이기위해 적절한 3개 값중 중간 값을 pivot으로 사용
    numOper = 0
    sample = [] #추출된 sample들 , index와 값을 하나로 묶음
    sample.append((0,arr[left])) #처음
    sample.append(((left+right)//2,arr[(left+right)//2])) #중간
    sample.append((right,arr[right])) #끝

    #값에 따라 정렬 한다. 버블 소트 이용
    for end in range(len(sample)-1,0,-1):
      for i in range(0,end,1):
        numOper+=1
        if(sample[i][1]>sample[i+1][1]):
          (sample[i],sample[i+1]) = (sample[i+1],sample[i]) #swap

    return numOper #계산 수를 반환


  def partition(self, arr, left, right): #퀵 정렬을 위한 partition함수 pivot을 뽑아 그 값을 제자리에 위치시킨후 위치 시킨 자리 인덱스를 반환
    numOper = 0 #시행 연산 수
    numOper += self.selectPivot(arr,left,right)#적절한 pivot을 맨 왼쪽으로 위치시킴

    low = left+1 #그 옆에서 부터 low시작
    high = right
    
    while(True):

      #low가 마지막에서 두번째 인덱스 이내 이고 arr[low]가 pivot값 보다 작거나 같으면 계속 왼쪽으로 이동
      while(low < right and arr[low] <= arr[left]): 
        low+=1 #low를 외쪽으로 이동시킴
        numOper+=1
      numOper+=1

      #high가 pivot 보다 오른쪽 이고 arr[high]가 pivot값 보다 크거나 같으면 오른쪽으로 이동 
      while(left < high and arr[left] <= arr[high]):#최종적으로 high가 pivot을 가리킬 수도 있음
        high-=1 #high를 오른쪽으로 이동시킴
        numOper +=1
      numOper+=1

      numOper+=1
      if(high<=low):#교차되거나 같을때
        break
      else:
        (arr[low],arr[high]) = (arr[high],arr[low]) #swap

    #high에 저장된 값이 우선순위가 pivot값과 같거나 빠르니 맨 왼쪽으로 보낼수 있음
    (arr[high],arr[left]) = (arr[left],arr[high]) #arr[high]와 pivot swap

    return high,numOper #pivot이들어간 index 반환,계산 수 반환


  #pivot을 그냥 맨 왼쪽으로 잡는 퀵소트
  def quickSort2(self, arr, left, right):

    if(left>=right):#종결 조건
      return 0
    else:
      pivotIdx, numOper = self.partition2(arr,left,right)

      numOper += self.quickSort2(arr,left,pivotIdx-1)#왼쪽 정렬
      numOper += self.quickSort2(arr,pivotIdx+1,right)#오른쪽 정렬

      return numOper


  def partition2(self, arr, left, right): #퀵 정렬을 위한 partition함수 pivot을 뽑아 그 값을 제자리에 위치시킨후 위치 시킨 자리 인덱스를 반환
    numOper = 0 #시행 연산 수
    low = left+1 #그 옆에서 부터 low시작
    high = right
    
    while(True):

      #low가 마지막에서 두번째 인덱스 이내 이고 arr[low]가 pivot값 보다 작거나 같으면 계속 왼쪽으로 이동
      while(low < right and arr[low] <= arr[left]): 
        low+=1 #low를 외쪽으로 이동시킴
        numOper+=1
      numOper+=1

      #high가 pivot 보다 오른쪽 이고 arr[high]가 pivot값 보다 크거나 같으면 오른쪽으로 이동 
      while(left < high and arr[left] <= arr[high]):#최종적으로 high가 pivot을 가리킬 수도 있음
        high-=1 #high를 오른쪽으로 이동시킴
        numOper +=1
      numOper+=1

      numOper+=1
      if(high<=low):#교차되거나 같을때
        break
      else:
        (arr[low],arr[high]) = (arr[high],arr[low]) #swap

    #high에 저장된 값이 우선순위가 pivot값과 같거나 빠르니 맨 왼쪽으로 보낼수 있음
    (arr[high],arr[left]) = (arr[left],arr[high]) #arr[high]와 pivot swap

    return high,numOper #pivot이들어간 index 반환,계산 수 반환


  def partiotion3(self,arr,low, high):
    numOper = 0

    pivot = arr[high] #high의 값을 pivot으로
    smallerIdx = low - 1 #pivot 보다 작은 것들이 차지한 자리

    for cur in range(low,high):
      numOper +=1
      if(arr[cur] <= pivot):#작거나 같은 것을 찾으면
        smallerIdx += 1 #smaller
        (arr[smallerIdx],arr[cur]) = (arr[cur],arr[smallerIdx]) #swap 작거나 같은 값을 채워 넣음
    
    (arr[smallerIdx+1], arr[high]) = (arr[high], arr[smallerIdx+1]) #swap
    
    return smallerIdx+1,numOper

  def quickSort3(self,arr,low,high):
    if(high<=low):#종결 조건
      return 0

    pivotIdx, numOper = self.partiotion3(arr, low, high)
    numOper += self.quickSort3(arr,low,pivotIdx-1)
    numOper += self.quickSort3(arr,pivotIdx+1,high)

    return numOper
```
